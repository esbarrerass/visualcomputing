Load model 1 and 2

```python
import trimesh

loaded_1 = trimesh.load('models/street_rat_4k.gltf') 
loaded_2 = trimesh.load('models/concrete_cat_statue_4k.gltf') 

if isinstance(loaded_1, trimesh.Scene):
    mesh_1 = trimesh.util.concatenate(
        tuple(loaded_1.geometry.values())
    )
else:
    mesh_1 = loaded_1

if isinstance(loaded_2, trimesh.Scene):
    mesh_2 = trimesh.util.concatenate(
        tuple(loaded_2.geometry.values())
    )
else:
    mesh_2 = loaded_2
```

Count faces, edges and vertixes


```python
num_vertices_1 = len(mesh_1.vertices)
num_faces_1 = len(mesh_1.faces)
num_edges_1 = len(mesh_1.edges_unique)

num_vertices_2 = len(mesh_2.vertices)
num_faces_2 = len(mesh_2.faces)
num_edges_2 = len(mesh_2.edges_unique)

print("Vertices model 1:", num_vertices_1)
print("Edges model 1:", num_edges_1)
print("Faces model 1:", num_faces_1)

print("Vertices model 2:", num_vertices_2)
print("Edges model 2:", num_edges_2)
print("Faces model 2:", num_faces_2)
```

Visualizing models 1

```python
from vedo import Mesh, Points, Plotter

vedo_mesh_1 = Mesh(mesh_1).c("lightblue").alpha(0.6).lighting("plastic")

vertices_1 = Points(mesh_1.vertices, r=3, c="red")

edges_lines_1 = vedo_mesh_1.boundaries().lw(2).c("green")

plt_1 = Plotter()

plt_1.camera.Azimuth(45)
plt_1.camera.Elevation(30)

plt_1.show(
    vedo_mesh_1,
    vertices_1,
    edges_lines_1,
    axes=1
)
```

Visualization model 2

```python
vedo_mesh_2 = Mesh(mesh_2).c("lightblue").alpha(0.6).lighting("plastic")

vertices_2 = Points(mesh_2.vertices, r=3, c="red")

edges_lines_2 = vedo_mesh_2.boundaries().lw(2).c("green")

plt_2 = Plotter()

plt_2.camera.Azimuth(45)
plt_2.camera.Elevation(30)

plt_2.show(
    vedo_mesh_2,
    vertices_2,
    edges_lines_2,
    axes=1
)
```
